# 1. Persiapan Data

```python
import pandas as pd

df = pd.read_csv('music_streaming_habits_2026.csv')
df.info()
df.head(10)
```


# 2. Analisis data

### 1. Analisis Keterikatan Pendengar Antar Platform Streaming Musik

```python
import matplotlib.pyplot as plt

platform_minutes = df.groupby('platform')['daily_listening_minutes'].mean()
platform_minutes.head(5)
plt.figure(figsize=(7, 5))

plt.bar(
    platform_minutes.index,
    platform_minutes.values,
    width=0.4
)

# 3. KUSTOMISASI GRAFIS
plt.title('Perbandingan Rata-rata Durasi Mendengarkan Harian', fontsize=12, fontweight='bold', pad=15)
plt.xlabel('Platform Musik', fontsize=10)
plt.ylabel('Durasi (Menit / Hari)', fontsize=10)

# TRICK: Grid hanya untuk sumbu Y (horizontal) agar grafik batang terlihat rapi dan tidak penuh tiang
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Tampilkan grafik
plt.tight_layout()
plt.show()
```

### 2. Pemetaan Preferensi Musik Global (Top 10 Artis Terpopuler)

```python
artis_favorit = df['top_artist'].value_counts().head(10)

plt.figure(figsize=(12, 9))
plt.barh(artis_favorit.index,
         artis_favorit.values
         )

plt.title('Top 10 Artis Yang Sering Didengarkan', fontsize=14, fontweight='bold')
plt.ylabel('Artist', fontsize = 12)
plt.xlabel('Akun Yang Mendengarkan', fontsize = 12)
plt.grid(axis='x', linestyle='--', alpha=0.5, color='black')
plt.tight_layout()
plt.gca().invert_yaxis()
plt.show()
```

```python
mood = df['top_mood'].value_counts()

plt.pie(
    mood.values,
    labels=mood.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#99ffff', '#ffe57f']
)

plt.title('Proporsi Top Mood Pendengar Musik 2026', fontsize=14, fontweight='bold', pad=20)
plt.show()
```
