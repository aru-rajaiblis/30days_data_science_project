# 1. Persiapan

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("AI_Impact_on_Jobs_2030.csv")
df.info()
```

# 2. Menjawab Pertanyaan Yang Sudah Dirumuskan

### 1. 10 Pekerjaan dengan Risiko AI Tertinggi

```python
(
    df.groupby("Job_Title")["AI_Replacement_Risk"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .sort_values(ascending=True)
    .plot(kind="barh", figsize=(10,5), color='crimson')
)

plt.title('Top 10 Pekerjaan Dengan Yang berpotensi Diganti AI', fontsize=12, fontweight='bold', pad = 15)
plt.xlabel('Rata-rata Skor Risiko AI', fontsize=12)
plt.ylabel('Pekerjaan', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7, color='black')
plt.xticks(rotation=40, ha='right')

plt.tight_layout()
plt.show()
```

### 2. Apakah industri  tertentu secara umum lebih rentan terhadap risiko AI dibandingkan industri lainnya?

```python
(
    df.groupby("Industry")["AI_Replacement_Risk"]
    .mean()
    .head(10)
    .sort_values(ascending=True)
    .plot(kind="barh", figsize=(10,5), color='Orange')
)

plt.title('Rata-rata Risiko Penggantian AI Berdasarkan Industri')
plt.xlabel('Rata-rata Skor Risiko AI')
plt.ylabel('Industri')
plt.grid(axis='x', linestyle='--', alpha=0.7, color='gray')
plt.xticks(rotation=40, ha='right')

plt.tight_layout()
plt.show()
```

### 3. Apakah pekerjaan dengan risiko AI tinggi cenderung memiliki rata-rata gaji yang lebih rendah, atau justru sebaliknya?

```python
(
    df.plot(kind="scatter",
            x='AI_Replacement_Risk',
            y='Average_Salary_USD',
            alpha=0.7,
            figsize=(10,5),
            color='teal')
)

plt.title('ScatterPlot Resiko AI Terhadap Pekerja Dengan Gaji', fontsize=12, fontweight='bold', pad = 15)
plt.xlabel('Rata-rata Skor Risiko AI', fontsize=12)
plt.ylabel('Gaji Pekerja USD', fontsize=12)
plt.grid(linestyle='--', alpha=0.7, color='gray')
plt.xticks(rotation=40, ha='right')

plt.tight_layout()
plt.show()
```

### 4. Di kelompok industri mana yang pekerjanya paling dinilai butuh meningkatkan keterampilan baru?

```python
(
    df[df['Upskilling_Needed'] == 'Yes']
    .groupby('Industry')
    .size()
    .sort_values(ascending=True)
    .plot(kind='barh', color='mediumseagreen', edgecolor='black', figsize=(10, 5))
)

plt.title('Industri Yang Membutuhkan Keterampilan Baru', fontsize=12, fontweight='bold', pad = 15)
plt.xlabel('Jumlah Pekerja yang Butuh Upskilling (Orang)')
plt.ylabel('Industri', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7, color='gray')
plt.xticks(rotation=40, ha='right')

plt.tight_layout()
plt.show()
```
